```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants
h = 6.626e-34
v_0 = 3.29e15 
n_H = 1e6
alpha = {"5000": 4.54e-19, "10000": 2.59e-19, "20000": 2.52e-19}

def quasar_function(v):
    I = ((1e24 * v_0**1.4) / (h)) * v**(-2.4) * np.exp(-(v) / (10 * v_0))
    return I

def quasar_integral(v):
    integral_sum = 0
    for i in range(1, len(v)):
        integral_sum += quasar_function(v[i]) * (v[i] - v[i-1])
    return integral_sum

def quasar_stromgren(Q_val):
    radius = ((3 * Q_val) / (4 * np.pi * n_H**2 * alpha['10000']))**(1/3)
    return radius

v = np.linspace(v_0, 1e18, 1000000)
Q = quasar_integral(v)

print(f'radius = {quasar_stromgren(Q)}')
```
